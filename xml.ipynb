```python
import re
import os
from bs4 import BeautifulSoup
import pandas as pd
```

```python
rootdir ='C:\\Users\\li\\Downloads\\project\\talkback-FUSION-201811072237-gdcpl4173\\fusion\\logs\\server\\'
filename = []
def listdir(path, list_name):
    for file in os.listdir(path):
        file_path = os.path.join(path, file)
        if os.path.isdir(file_path):
            listdir(file_path, list_name)
        elif os.path.splitext(file_path)[1]=='.xml':
            list_name.append(file_path)
listdir(rootdir,filename)
```

```python
def find(lis,whole):
    pattern = re.compile("\>(.*?)\<")
    check_list = []
    taskid_list = lis.find_all("taskid")
    taskid = pattern.findall(str(taskid_list))
    check_list.append(taskid[0])

    timecreated_list = lis.find_all("timecreated")
    timecreated = pattern.findall(str(timecreated_list))
    check_list.append(timecreated[0])

    creatornodeid_list = lis.find_all("creatornodeid")
    creatornodeid = pattern.findall(str(creatornodeid_list))
    check_list.append(creatornodeid[0])

    timeupdated_list = lis.find_all("timeupdated")
    timeupdated = pattern.findall(str(timeupdated_list))
    check_list.append(timeupdated[0])

    taskidentity_list = lis.find_all("taskidentity")
    taskidentity = pattern.findall(str(taskidentity_list))
    check_list.append(taskidentity[0])

    uri_list = lis.find_all("uri")
    uri = pattern.findall(str(uri_list))
    uri_s = ''
    for uri_ in uri:
        if (uri_ != '') & (uri_ != ', '):
            uri_s += '    uri: '
            uri_s += uri_
    check_list.append(uri_s)

    nodeidentity_list = lis.find_all("nodeidentity")
    nodeidentity1 = pattern.findall(str(nodeidentity_list[0]))
    check_list.append(nodeidentity1[0])
    nodeidentity2 = pattern.findall(str(nodeidentity_list[1]))
    check_list.append(nodeidentity2[0])

    checkTask_list = lis.find_all("com.wandisco.fs.server.consistencycheck.consistencychecktype")
    check_list.append(checkTask_list)

    affectedzones_list = lis.find_all("affectedzones")
    check_list.append(affectedzones_list[0])

    zoneStates_list = lis.find_all("zoneStates")
    check_list.append(zoneStates_list)
    whole.append(check_list)
    return whole
```

```python
whole = []
for i in range(len(filename)):
    file_dir = filename[i]
    data = open(file_dir).read()
    soup = BeautifulSoup(data,'lxml')
    checkTask = soup.find_all("com.wandisco.fs.server.consistencycheck.consistencychecktask")
    if checkTask != []:
        for check in checkTask:
            whole = find(check,whole)
```

```python
# consistencycheck
data= pd.DataFrame(whole)
data.rename(columns={0:'taskid',1:'timecreated',2:'creatornodeid',3:'timeupdated',4:'taskidentity',5:'uri',6:'nodeidentity1',7:'nodeidentity2',8:'checkTask',9:'affectedzones',10:'checkTask',11:'zoneStates'},inplace=True)
data
```
