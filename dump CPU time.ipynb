### 2.compare CPU time for each of these threads across each hourly thread-dump logFile

in a busy system, threads should have ever increasing CPU time, so I get the DSM and cpu times

```python
time_ptn = re.compile('OutputProposalSequence:\[\[.*?\]\[(.*?)\]\]\-0:\s+(\d+)\s+\d+\s+\d+\s+[A-Z]+')
```

```python
data = ''
for name in filename:
    direction = rootdir + name
    data += open(direction).read()
```

```python
DSM_time = time_ptn.findall(data)
```

```python
DSM = {}
for i in range(len(DSM_time)):
    dsmId = DSM_time[i][0]
    cputime = int(DSM_time[i][1])
    if dsmId in DSM:
        DSM[dsmId].append(cputime)
    else:
        DSM[dsmId] = []
        DSM[dsmId].append(cputime)
```

```python
strange_dsm = []
for dsm, time in DSM.items():
    for i in range(len(time)-1):
        if time[i+1] - time[i] < 0:
            strange_dsm.append(dsm)
```

but seems like all threads are increasing CPU time for each hour
